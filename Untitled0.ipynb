## Data loading:
Load the Alzheimer's disease dataset.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('alzheimers_disease_data.csv')
display(df.head())
```

## Data Cleaning



```python
print(f"Shape of the DataFrame: {df.shape}")
```

```python
print("Basic Information about the Data:")
df.info()
```

```python
print("Number of Missing Values in Each Column:")
print(df.isnull().sum())
```

```python
print("\nDescriptive Statistics:")
display(df.describe())
```

```python
print("Number of Duplicate values in the column:")
print(df.duplicated().sum())
```

```python
df_reverse = df.copy()

gender_mapping = {0: 'Male', 1: 'Female'}
ethnicity_mapping = {0: 'Caucasian', 1: 'African American', 2: 'Asian', 3: 'Other'}
education_mapping = {0: 'None', 1: 'High School', 2: 'Bachelor\'s', 3: 'Higher'}
binary_mapping = {0: 'No', 1: 'Yes'}

df_reverse['Gender'] = df_reverse['Gender'].map(gender_mapping)
df_reverse['Ethnicity'] = df_reverse['Ethnicity'].map(ethnicity_mapping)
df_reverse['EducationLevel'] = df_reverse['EducationLevel'].map(education_mapping)
df_reverse['Smoking'] = df_reverse['Smoking'].map(binary_mapping)
df_reverse['FamilyHistoryAlzheimers'] = df_reverse['FamilyHistoryAlzheimers'].map(binary_mapping)
df_reverse['CardiovascularDisease'] = df_reverse['CardiovascularDisease'].map(binary_mapping)
df_reverse['Diabetes'] = df_reverse['Diabetes'].map(binary_mapping)
df_reverse['Depression'] = df_reverse['Depression'].map(binary_mapping)
df_reverse['HeadInjury'] = df_reverse['HeadInjury'].map(binary_mapping)
df_reverse['Hypertension'] = df_reverse['Hypertension'].map(binary_mapping)
df_reverse['MemoryComplaints'] = df_reverse['MemoryComplaints'].map(binary_mapping)
df_reverse['BehavioralProblems'] = df_reverse['BehavioralProblems'].map(binary_mapping)
df_reverse['Confusion'] = df_reverse['Confusion'].map(binary_mapping)
df_reverse['Disorientation'] = df_reverse['Disorientation'].map(binary_mapping)
df_reverse['PersonalityChanges'] = df_reverse['PersonalityChanges'].map(binary_mapping)
df_reverse['DifficultyCompletingTasks'] = df_reverse['DifficultyCompletingTasks'].map(binary_mapping)
df_reverse['Forgetfulness'] = df_reverse['Forgetfulness'].map(binary_mapping)
df_reverse['Diagnosis'] = df_reverse['Diagnosis'].map(binary_mapping)


display(df_reverse.head())
```

```python
if 'DoctorInCharge' in df.columns:
    df.drop('DoctorInCharge', inplace=True, axis=1)
    print("Column 'DoctorInCharge' removed successfully.")
else:
    print("Column 'DoctorInCharge' not found in the DataFrame.")
```

#Data Exploration

```python
df['Diagnosis'].value_counts().plot(kind='pie',autopct = '%.1f%%', figsize=(5,5));
```

```python
num_cols = df.select_dtypes('int64').columns.to_list()
num_cols.remove('Age')
num_cols.remove('Diagnosis')
num_cols.remove('SystolicBP')
num_cols.remove('DiastolicBP')
num_cols.remove('PatientID')

fig, axes = plt.subplots(9, 2, figsize=(14, 36))  # Adjust figsize for better spacing

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Plot each column
for idx, column in enumerate(num_cols):
    sns.countplot(x=df[column], hue=df['Diagnosis'], ax=axes[idx])
    axes[idx].set_title(f'Countplot of {column}')

# Remove any empty subplots
for i in range(len(num_cols), len(axes)):
    fig.delaxes(axes[i])

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```


#Research question 1

Will cognitive assessments (MMSE, Memory Complaints) help predict Alzheimer’s before clinical diagnosis?


```python
df_cognitive = df[['MMSE', 'MemoryComplaints','FunctionalAssessment', 'BehavioralProblems', 'Diagnosis']]
correlation = df_cognitive.corr()
print(correlation)

sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation between Cognitive Assessments and Diagnosis')
plt.show()
```

There is not much correlation with cognitive assesments.

#Research question 2
How does lifestyle factor correlate with Alzheimer’s diagnosis?

```python
df_lifestyle = df[['BMI','Smoking','AlcoholConsumption','PhysicalActivity','DietQuality','SleepQuality', 'Diagnosis']]
correlation = df_lifestyle.corr()
print(correlation)

sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation between Cognitive Assessments and Diagnosis')
plt.show()
```

There is not much correlation with lifestyle.
